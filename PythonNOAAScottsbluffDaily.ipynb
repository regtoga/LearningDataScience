```python
import pandas as pd
import matplotlib.pyplot as plt

#Read the data from the CSV file "3671811.csv" into a Dataframe named "noaa"
noaa = pd.read_csv(
    "https://raw.githubusercontent.com/regtoga/LearningDataScience/main/3671811.csv"
)

#Display the first and last few rows of the DataFrame "noaa".
#This gives you an idea about the types of data
print(noaa.head())
print(noaa.tail())

#Extract the 'TMAX' and 'TMIN' colums from the DataFrame "noaa"
#assign to variables
temp_max = noaa['TMAX']
temp_min = noaa['TMIN']

#Create a scatter plot with "temp_min" as the x-values,
#"temp_max" as the y-valuesl, marker size of 100,
#filled markers in blue color with transparency set to 0.5.
plt.scatter(temp_min, temp_max, s=100, c="blue", alpha=0.5)

#Display the grid on the plot
plt.grid(True)

#Set the title of the plot to "Min Max Temperatures Jan-Mar 2024".
plt.title("Min Max Temperatures April 2023-24")

#Set the label for the x-axis as "Minimum Temperature"
plt.xlabel("Minimum Temperature")

#Set the label for the y-axis as "Maximum Temperature".
plt.ylabel("Maximum Temperature")

#Display the plot.
plt.show()
```

```python
# Line plot showing TMIN and TMAX temperatures on the same plot
plt.figure(figsize=(10, 5))
plt.plot(noaa["DATE"], noaa["TMIN"], label="TMIN")
plt.plot(noaa["DATE"], noaa["TMAX"], label="TMAX")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("TMIN and TMAX Temperatures")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Line plot showing TAVG
noaa["TAVG"] = (noaa["TMIN"] + noaa["TMAX"]) / 2

plt.figure(figsize=(10, 5))
plt.plot(noaa["DATE"], noaa["TAVG"], label="TAVG", color="green")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("TAVG Temperatures")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar plot showing TAVG
plt.figure(figsize=(10, 5))
plt.bar(noaa["DATE"], noaa["TAVG"], color="orange")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("TAVG Temperatures")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
